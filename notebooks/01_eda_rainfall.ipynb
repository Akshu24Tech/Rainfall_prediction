```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data/raw/rainfall in india 1901-2015.csv')
print(df.head())

# check missing values
print(df.isnull().sum())

sns.lineplot(x='YEAR', y='ANNUAL', data=df)
plt.title("Annual Rainfall Over Years")
plt.show()
```
